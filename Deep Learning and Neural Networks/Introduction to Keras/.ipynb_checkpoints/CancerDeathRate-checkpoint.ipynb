```python
#import library
from IPython.display import display, HTML
import pandas as pd
import numpy as np
import pickle
import re
from IPython.display import display, HTML
```

```python
asian_country = {}
```

```python
#Fetch Asian countried from www.worldometers.info
country= pd.read_html('https://www.worldometers.info/geography/how-many-countries-in-asia/')
```

```python
asian_country = country[0]['Country']
```

```python
asian_country
```

```python
#Featch data of deaths  by cause
df=pd.read_csv("annual-number-of-deaths-by-cause.csv")
```

```python
df
```

```python
#We are interested in cancer data so lets filter it
df_cancer=df[['Entity','Year','Cancers (deaths)']]
```

```python
df_cancer
```

```python
df_year = df_cancer.pivot_table('Cancers (deaths)', ['Entity'], 'Year')
```

```python
df_year
```

```python
df_year.reset_index(inplace=True)
```

```python
#filter data for asian countries
df_year = df_year[df_year.Entity.isin(asian_country)]
```

```python
#Data is redy for visulization lest export it into csv file
df_year.to_csv (r'C:\Users\z0040vsk\Downloads\export_dataframe.csv', index = False, header=True)
```
