# Linear Regression - Assignment


### Problem Statement


A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts. 

They have contracted an automobile consulting company to understand the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting the pricing of cars in the American market, since those may be very different from the Chinese market. The company wants to know:

    1) Which variables are significant in predicting the price of a car.
    2) How well those variables describe the price of a car.

## Step 1: Importing and Understanding Data

**Import necessary libraries** 

```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.metrics import r2_score

# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

import statsmodels.api as sm

# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

### Fixing Rows and Columns

```python
#Delete car_ID column as it just index column
carInfo.drop(['car_ID'], axis =1, inplace = True)
```

```python
#Lets extract the company name from column CarName
carInfo['CarName'] = carInfo['CarName'].str.replace('-',' ')
carInfo['car_company'] = carInfo.CarName.map(lambda x: x.split(" ", 1)[0])
carInfo.car_company.unique()
```

Correct the car company names, the names spelling looks incorrect.

```python
carInfo['car_company'] = carInfo['car_company'].str.replace('vw','volkswagen')
carInfo['car_company'] = carInfo['car_company'].str.replace('vokswagen','volkswagen')
carInfo['car_company'] = carInfo['car_company'].str.replace('toyouta','toyota')
carInfo['car_company'] = carInfo['car_company'].str.replace('porcshce','porsche')
carInfo['car_company'] = carInfo['car_company'].str.replace('maxda','mazda')
carInfo['car_company'] = carInfo['car_company'].str.replace('Nissan','nissan')
carInfo.car_company.unique()
```

```python
#Drop CarName column, as useful data is extracted in new car_company column
carInfo.drop(['CarName'], axis =1, inplace = True)
```

#### Lets delete data from categorial column which has one or two row of information, otherwise it will add extra dummy variable column

```python
#Delete row of engintype dohcv
carInfo.drop(carInfo.loc[carInfo['enginetype'] == 'dohcv'].index, inplace =True)

#Delete row of cylinder number three and twelve
carInfo.drop(carInfo.loc[carInfo['cylindernumber'] == 'three'].index, inplace = True)
carInfo.drop(carInfo.loc[carInfo['cylindernumber'] == 'twelve'].index, inplace = True)

#Delete row of fuelsystem mfi
carInfo.drop(carInfo.loc[carInfo['fuelsystem'] == 'mfi'].index, inplace = True)
```

## Step 3: Data Preparation

**To fit a regression line, we would need numerical values and not string. Hence, we need to convert them to 1s and 0s**
<ul>
  <li>Map doornumber and cylindernumber string values with respective its numeric values.</li>
  <li>Map aspiration with one  as std and zero as turbo.</li>
  <li>Map enginelocation with  one as front and zero as rear.</li>
</ul>

```python
carInfo['doornumber'] = carInfo['doornumber'].map({'two': 2, 'four': 4})
carInfo['cylindernumber'] = carInfo['cylindernumber'].map({'two': 2,'four': 4,'five': 5, 'six': 6, 'eight':8})
carInfo['fueltype'] = carInfo['fueltype'].map({'diesel': 1, 'gas': 0})
carInfo['aspiration'] = carInfo['aspiration'].map({'std': 1, 'turbo': 0})
carInfo['enginelocation'] = carInfo['enginelocation'].map({'front': 1, 'rear': 0})
```

```python
carInfo.info()
```

### Dummy Variables

**Lets genrate dummy values for categorical columns car_company, carbody, drivewheel, enginetype, fuelsystem.**

```python
carInfo = pd.get_dummies(carInfo)
```

```python
carInfo.info()
```

```python
carInfo.head()
```

## Step 4: Splitting the Data into Training and Testing Sets

```python
np.random.seed(0)
car_train, car_test = train_test_split(carInfo, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescaling the Features

**To avoid the huge differntiation between coefficients which will obtain by fitting the regression model we will use Min-Max scaling**  

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['symboling', 'wheelbase', 'carheight', 'cylindernumber','enginesize', 'boreratio', 'stroke','compressionratio','horsepower','peakrpm','highwaympg','price']
car_train[num_vars] = scaler.fit_transform(car_train[num_vars])
```

```python
car_train.head()
```

```python
car_train.describe()
```

### Dividing into X and Y sets for the model building

```python
y_train = car_train.pop('price')
X_train = car_train
```

## Step 5: Building a linear model

### RFE
**Recursive feature elimination**

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)
rfe = rfe.fit(X_train, y_train)
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
#The column which needs to be drop
col = X_train.columns[~rfe.support_]
col
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train.drop(col,1)
X_train_rfe.columns
```

```python
# Adding a constant variable 
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model
lm1 = sm.OLS(y_train, X_train_rfe).fit()

#Let's see the summary of our linear model
print(lm1.summary())
```

```python
#Fnction to calculate VIF
def vif_cal(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2) 
    return vif.sort_values(by = "VIF", ascending = False)
```

```python
#Check VIF
vif_cal(X=X_train_rfe)
```

### Dropping the variable and updating the model

From the summary and VIF dataframe, cylindernumber looks insignificant. it has very high P-value 0.258. 
Drop cylindernumber

```python
X_train_one = X_train_rfe.drop(["cylindernumber"], axis = 1)
```

Rebuilding the model without cylindernumber

```python
X_train_one =  sm.add_constant(X_train_one)

# Running the linear model
lm2 = sm.OLS(y_train,X_train_one).fit()   
print(lm2.summary())
```

```python
#Check VIF
vif_cal(X=X_train_one)
```

```python
X_train_two = X_train_one.drop(["car_company_alfa"], axis = 1)
```

Rebuilding the model without car_company_alfa

```python
X_train_two =  sm.add_constant(X_train_two)

# Running the linear model
lm3 = sm.OLS(y_train,X_train_two).fit()   
print(lm3.summary())
```

From the summary, enginetype_ohcv looks insignificant. it has very high P-value 0.060. Drop enginetype_ohcv

```python
X_train_three = X_train_two.drop(["enginetype_ohcv"], axis = 1)
```

Rebuilding the model without enginetype_ohcv

```python
X_train_three =  sm.add_constant(X_train_three)

# Running the linear model
lm4 = sm.OLS(y_train,X_train_three).fit()   
print(lm4.summary())
```

```python
#Check VIF
vif_cal(X=X_train_three)
```

From VIF dataframe, enginesize looks insignificant. It VIF value is grater than 5. Drop enginesize

Rebuilding the model without peakrpm

```python
#Add Constant 
X_train_five =  sm.add_constant(X_train_five)

# Running the linear model
lm6 = sm.OLS(y_train,X_train_five).fit()   
print(lm6.summary())
```

From the summary, stroke looks insignificant. it has very high P-value 0.115. Drop stroke

```python
X_train_six = X_train_five.drop(["stroke"], axis = 1)
```

Rebuilding the model without stroke

```python
#Add Constant 
X_train_six =  sm.add_constant(X_train_six)

# Running the linear model
lm7 = sm.OLS(y_train,X_train_six).fit()   
print(lm7.summary())
```

## Step 7: Residual Analysis of the train data
